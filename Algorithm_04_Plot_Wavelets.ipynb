```python
#---------------------------------- Flatten Input a 2D-Wavelet Array tp 1D Vector in order to enter the ML
def One_Dimention_Tranformation(img):
    nx, ny = img.shape
    img_1d = img.reshape((nx*ny))
    return(img_1d)

#---------------------------- This function takes one list of 2D images/arrays and returns one 2D image/array which is the representative of this list (with the highest correlation to all others)
def Wavelet_Correlation(images_list):
    images_list_resized =  Resize_list_of_Imges(images_list,images_list.shape[1]/2,images_list.shape[2]/5) #------- Make the IMGS smaller for quicker execution
    number_of_images = len(images_list_resized) 
    images_list_1d = [] #-------------- Flatten them so to calculate th PCC
    for i in range(0,number_of_images):
        one_dimention_data =  One_Dimention_Tranformation(images_list_resized[i])
        images_list_1d.append(one_dimention_data)
    images_list_1d = np.array(images_list_1d)
    #---------------------------------- with the following code we create the correlation PCC Matrix
    adj_mtrx = np.zeros((number_of_images, number_of_images))
    for i in tqdm(range(0,number_of_images)): 
        for j in range(0,number_of_images):
            max_pcc_valuea, _ = pearsonr(images_list_1d[i],images_list_1d[j]) #---------------------------- PCC Linear Correlation
            adj_mtrx[i,j] = max_pcc_valuea
    return(adj_mtrx)

wavelet_correlation_array = Wavelet_Correlation(burst_data)
#for i in range(0,wavelet_correlation_array.shape[0]):
#    for j in range(0,wavelet_correlation_array.shape[1]):
#        if(wavelet_correlation_array[i,j]>=0.70):
#            wavelet_correlation_array[i,j]=0.9

plt.imshow(wavelet_correlation_array, cmap='seismic', aspect='auto',vmin=0.3,vmax=1)
cbar = plt.colorbar()
plt.title("Linear Correlation between Class Representatives")
plt.ylabel("Wavelets")
plt.xlabel("Wavelets")
cbar.set_label('1 --> High Correlation, 0 --> Low correlation)',size=10)
plt.show()
```

```python
burst_data = wavelet_data[0]
for i in range(1,40):
    burst_data = np.append(burst_data,wavelet_data[i],0)
print(burst_data.shape)
      
wavelet_correlation_array = Wavelet_Correlation(burst_data)

high_correlation_counter = np.zeros((wavelet_correlation_array.shape[0]))
for i in range(0,wavelet_correlation_array.shape[0]):
    for j in range(0,wavelet_correlation_array.shape[1]):
        if(wavelet_correlation_array[i,j]>=0.75):
            high_correlation_counter[i] +=1

counter_low_correlation_10 = 0
counter_low_correlation_100 = 0
for i in range(0,len(high_correlation_counter)):
    if(high_correlation_counter[i]<=10):
        counter_low_correlation_10+=1
    if(high_correlation_counter[i]<=100):
        counter_low_correlation_100+=1
print(counter_low_correlation_10/len(high_correlation_counter))
print(counter_low_correlation_100/len(high_correlation_counter))
```
