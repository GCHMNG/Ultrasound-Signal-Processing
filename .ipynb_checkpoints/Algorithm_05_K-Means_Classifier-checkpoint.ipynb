```python
#---------------------------------- KMeans classification Tensoflow
from sklearn import cluster
from collections import defaultdict
from matplotlib import cm
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn.metrics.cluster import adjusted_rand_score

#---------------------------------- Flatten Input a 2D-Wavelet Array tp 1D Vector in order to enter the ML
def One_Dimention_Tranformation(img):
    nx, ny = img.shape
    img_1d = img.reshape((nx*ny))
    return(img_1d)

#---------------------------------- Make wavelets an one list from list of bursts of wavelets
def Continues_1D_List(wavelet_data):
    wavelet_data_list = []
    for i in range(0,wavelet_data.shape[0]):
        for j in range(0,wavelet_data.shape[1]):
            frame =  wavelet_data[i,j]
            wavelet_data_list.append(frame)
    wavelet_data_list = np.array(wavelet_data_list)
    return (wavelet_data_list)

#---------------------------------- Short the Classes based on their location over time: class 0 left/early, class N right/late
def Shorting_ML_Classes(classes_number,kmeans_results):
    sums_indices_of_init_classes = np.zeros((classes_number))
    numb_indices_of_init_classes = np.zeros((classes_number))
    avrg_indices_of_init_classes = np.zeros((classes_number))
    for i in range(0,len(kmeans_results)):
        curr_class = kmeans_results[i]
        sums_indices_of_init_classes[curr_class] += i
        numb_indices_of_init_classes[curr_class] += 1
    for i in range(0,len(avrg_indices_of_init_classes)):
        avrg_indices_of_init_classes[i] = sums_indices_of_init_classes[i]/numb_indices_of_init_classes[i]
    #print("Avrg Init Indeces:",list(map(int, avrg_indices_of_init_classes)))
    sort_avrg_indices_of_init_classes = np.argsort(avrg_indices_of_init_classes)
    int_classes = list(range(0, classes_number))
    sorted_classes = np.zeros((classes_number))
    for i in range(0,len(int_classes)):
        sorted_classes[i] = int_classes[sort_avrg_indices_of_init_classes[i]]
    sorted_classes = list(map(int, sorted_classes))
    #print("Init Classes:",int_classes)
    #print("Sorted Classes",sorted_classes)
    dct = dict([(int(sorted_classes[0]), int(int_classes[0]))])
    for i in range(1,classes_number):
        dct[int(sorted_classes[i])] = int(int_classes[i])
    #print("Dictionary",dct)
    shorted_kmeans_results = [dct[k] for k in kmeans_results]    
    return (shorted_kmeans_results,dct)

#---------------------------------- Returns an Adjacency Matrix of the centers of all Classes
def ML_Center_Differences(classes_number,kmeans_centers,dct):
    kmeans_centers = np.array(kmeans_centers)
    print("Shape of Centers of all Classes:",kmeans_centers.shape)
    shorted_kmeans_centers =[]
    for key, val in dct.items(): # sort them based on the shorting of classes
        shorted_kmeans_centers.append(kmeans_centers[key])
    shorted_kmeans_centers = np.array(shorted_kmeans_centers)
    #---------------------------- this is the Adjacency Matrix of the centers
    distances_between_centers = np.zeros((classes_number, classes_number)) 
    for i in range(0,distances_between_centers.shape[0]):
        for j in range(0,distances_between_centers.shape[1]):
            distances_between_centers[i,j] = np.linalg.norm(shorted_kmeans_centers[i]-shorted_kmeans_centers[j])
    max_val = np.amax(distances_between_centers) # Here we make the Normalization - between 0-1
    for i in range(0,distances_between_centers.shape[0]):
        for j in range(0,distances_between_centers.shape[1]):
            distances_between_centers[i,j] = distances_between_centers[i,j] / max_val
    print("Shape of Adjacency Matrix of the Centers:",distances_between_centers.shape)
    return(distances_between_centers)
   
#---------------------------------- List of Lists of Wavelets, each list have all the wavelets from the same Class
def Lists_Wavelets_In_Same_Classes(classes_number,wavelet_data_list,shorted_kmeans_results):
    #---------------------------- Inicialy we make an empty list of lists with the appropriate size of classes
    Wavelets_classified = [[]] * classes_number
    for i in range(0,len(Wavelets_classified)): 
        Wavelets_classified[i] = Wavelets_classified[i] + list()
    #---------------------------- we place the 2D-Wavelets in this list of lists 
    for i in range(0,len(wavelet_data_list)):
        for j in range(0,classes_number):
            if(shorted_kmeans_results[i]==j): 
                Wavelets_classified[j].append(list(wavelet_data_list[i]))
    print("Place Wavelets to same Classes Lists --> Completed")
    #print("Classes Shapes:")
    #for i in range(0,classes_number):
    #    print("Class:",i,"--> Number of Wavelets:",np.array(Wavelets_classified[i]).shape[0])
    return(Wavelets_classified)

#---------------------------- This function takes one list of 2D images/arrays and returns one 2D image/array which is the representative of this list (with the highest correlation to all others)
def Class_Representative(images_list):
    images_list_resized =  Resize_list_of_Imges(images_list,images_list.shape[1]/4,images_list.shape[2]/4) #------- Make the IMGS smaller for quicker execution
    number_of_images = len(images_list_resized) 
    images_list_1d = [] #-------------- Flatten them so to calculate th PCC
    for i in range(0,number_of_images):
        one_dimention_data =  One_Dimention_Tranformation(images_list_resized[i])
        images_list_1d.append(one_dimention_data)
    images_list_1d = np.array(images_list_1d)
    #---------------------------------- with the following code we create the correlation PCC Matrix
    adj_mtrx = np.zeros((number_of_images, number_of_images))
    for i in tqdm(range(0,number_of_images)): 
        for j in range(0,number_of_images):
            max_pcc_valuea, _ = pearsonr(images_list_1d[i],images_list_1d[j]) #---------------------------- PCC Linear Correlation
            adj_mtrx[i,j] = max_pcc_valuea
    #---------------------------------- we find the average pcc of each image
    avg_pcc_of_each_image = np.zeros(number_of_images)
    for i in range(0,number_of_images): 
        sum_pcc_of_image = 0
        for j in range(0,number_of_images):
            sum_pcc_of_image += adj_mtrx[i,j]
        avg_pcc_of_each_image[i] = sum_pcc_of_image/number_of_images
    #---------------------------------- we select the image with the highest average PCC
    max_index = 0
    max_value = 0
    for i in range(0,number_of_images): 
        if(avg_pcc_of_each_image[i]>max_value):
            max_index = i
            max_value = avg_pcc_of_each_image[i]
    return(images_list[max_index])

#---------------------------------- This function takes all the lists of classifed wavelet images an returns a list of one representative image of each of those classes
def Classes_Representatives(classes_number,Wavelets_classified):
    Wavelets_Representatives_of_each_Class = []
    for i in range(0,classes_number):
        current_class_wavelet_images = np.array(Wavelets_classified[i])
        current_class_representative = Class_Representative(current_class_wavelet_images)
        Wavelets_Representatives_of_each_Class.append(current_class_representative)
    Wavelets_Representatives_of_each_Class = np.array(Wavelets_Representatives_of_each_Class)
    print("Classes Representative shape:",Wavelets_Representatives_of_each_Class.shape)
    return(Wavelets_Representatives_of_each_Class)

#---------------------------------- MAIN ML Classifier Function
def ML_Classifier(dataset_path,freqs_path,classes_number,n_init_val):
    #---------------------------- Load Wavelet data
    wavelet_trackdata = io.loadmat(dataset_path)
    wavelet_data = wavelet_trackdata['saved_data'];
    print("Number of Bursts in Dataset:",wavelet_data.shape[0])
    print("Number of Wavelets/Spikes per Burst:",wavelet_data.shape[1])
    print("Total Number of Wavelets in Dataset:",wavelet_data.shape[0]*wavelet_data.shape[1])
    #print("Wavelet 2D-Image with shape:",wavelet_data.shape[2:])     
    #---------------------------- Load Frequency data
    freqs_trackdata = io.loadmat(freqs_path)
    freqs_data = freqs_trackdata['saved_data'];
    freqs_data = freqs_data[0]
    #---------------------------- Make put all wavelets from lists of bursts to one continues list
    wavelet_data_list = Continues_1D_List(wavelet_data)
    print("Wavelets in one list:",wavelet_data_list.shape)
    #---------------------------- Normalize Data
    #max_val = np.amax(wavelet_data_list)
    #wavelet_data_list = np.divide(wavelet_data_list, max_val)
    #---------------------------- Flatten Wavelet data and put them in 1 continues list 
    wavelet_data_1d = []
    for i in range(0,wavelet_data_list.shape[0]):
        one_dimention_data =  One_Dimention_Tranformation(Resize_Img(wavelet_data_list[i],wavelet_data_list.shape[1]/2,wavelet_data_list.shape[2]/2)) # <--------RESIZE the 2D Array in half for quicker ML execution time
        wavelet_data_1d.append(one_dimention_data)
    wavelet_data_1d = np.array(wavelet_data_1d)
    print("Wavelet Flatten, from 2D shape:",wavelet_data_list.shape[1:],"--> to 1D shape:",wavelet_data_1d.shape[1],"*Resized")    
    #---------------------------- K-Means classification
    kmeans_algorithm = cluster.KMeans(n_clusters=classes_number, algorithm='full', n_init= n_init_val)
    # https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html 
    kmeans_algorithm.fit(wavelet_data_1d)
    kmeans_results = list(kmeans_algorithm.labels_)
    kmeans_centers = list(kmeans_algorithm.cluster_centers_)
    print("Classification Completed---------------------------------------------------")
    shorted_kmeans_results,dct = Shorting_ML_Classes(classes_number,kmeans_results)
    distances_between_centers = ML_Center_Differences(classes_number,kmeans_centers,dct)
    wavelet_data_classified = Lists_Wavelets_In_Same_Classes(classes_number,wavelet_data_list,shorted_kmeans_results)
    representative_wavelets = Classes_Representatives(classes_number,wavelet_data_classified)

    print("---------------------------End of Classification---------------------------")
    return(wavelet_data,freqs_data,shorted_kmeans_results,distances_between_centers,wavelet_data_classified,representative_wavelets)

def Plot_Classes_Properties(classes_labels,classes_center_distances):
    plt.figure(1, figsize=(12,10))
    plt.clf()
    ax = plt.subplot(221) #---------------------------------- Plot Classes over time 
    plt.plot(classes_labels,'o')
    plt.title("Distribution of classified Wavelets during the experiment")
    plt.ylabel("Classes")
    plt.xlabel("Wavelets/Spikes over time")
    plt.show
    ax = plt.subplot(222) #---------------------------------- Plot Adjacency Matrix of Centers
    plt.imshow(classes_center_distances, cmap='seismic', aspect='auto')
    cbar = plt.colorbar()
    plt.title("Distances Between the Centers of each Class")
    plt.ylabel("Classes Centers")
    plt.xlabel("Classes Centers")
    cbar.set_label('Distance of Classes Centers (1=furthest, 0=same spot)',size=10)
    plt.show()
    
#---------------------------- These functions we plot 1 CWT wavelet AND it is called only from the Plot_CTW_Rows_From_Saved_Wavelets. 
def Sub_Plot_Wavelet (ax,i,wavelet,freqs):
    plt.title("Class "+np.str(i))
    sample_Rate = 100000000  # in Hertz <-- (100MHz)
    period = 1/sample_Rate # in seconds
    period = period * 1000000 # in microseconds (μs)
    sample_Rate = sample_Rate / 1000000 # in MegaHz
    #ax = plt.gca()
    im = ax.imshow(wavelet, cmap='seismic', aspect='auto',vmin=0, vmax=1)  # seismic gray
    plt.colorbar(im)
    #---------------------------- Axis
    x_values = np.arange(0,200,35)
    y_values = np.arange(0,len(freqs),15)
    x_labels = []
    for i in x_values:
        x_labels.append(np.str("%.2f" % np.round_(i*period, 4)))
    #y_labels = []
    #for i in y_values:
    #    y_labels.append(np.str("%.2f" % np.round_(freqs[i], 4)))
    ax.set_xticks(x_values)
    ax.set_xticklabels(x_labels)
    ax.set_yticks(y_values)
    #ax.set_yticklabels( y_labels )
    plt.ylabel("Frequency - MegaHertz (MHz)")
    plt.xlabel("Time - microseconds (μs)")

#---------------------------- This function we plot rows of saved wavelets 
def Plot_Rows_of_Wavelets(wavelet_data,freqs,rows_num):
    print("Wavelets Representatives of each Class:")
    wavelets_num = len(wavelet_data)
    i = 0
    while (i < len(wavelet_data)):
        plt.figure(1, figsize=(40,6))
        plt.clf()
        for j in range(0,rows_num):
            if(i+j<wavelets_num):
                ax = plt.subplot(201+rows_num*10+j) #---------------------------------- PLOT
                Sub_Plot_Wavelet (ax,i+j,wavelet_data[i+j],freqs)
                plt.show
            else:
                break
        plt.show()
        i +=rows_num
```
