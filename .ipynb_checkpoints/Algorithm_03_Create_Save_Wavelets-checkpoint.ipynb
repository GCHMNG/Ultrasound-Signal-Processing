```python
#--------------------------------- ALL in one
#signal_list = [signal_o_1,signal_o_2,signal_o_3,signal_o_4,signal_o_5,signal_o_6,signal_o_7,signal_o_8,signal_c_1,signal_c_2,signal_c_3,signal_c_4,signal_c_5]
#images_list = [img_o_1,img_o_2,img_o_3,img_o_4,img_o_5,img_o_6,img_o_7,img_o_8,img_c_1,img_c_2,img_c_3,img_c_4,img_c_5]
#freqs_list =[freqs_o_1,freqs_o_2,freqs_o_3,freqs_o_4,freqs_o_5,freqs_o_6,freqs_o_7,freqs_o_8,freqs_c_1,freqs_c_2,freqs_c_3,freqs_c_4,freqs_c_5] 

signal_list = [signal_dp_1,signal_dp_2,signal_dp_3,signal_dp_4,signal_dp_5,signal_dp_6,signal_dp_7,signal_dp_8,signal_dp_9,signal_dp_10,signal_dp_11,signal_dp_12,signal_dp_13,signal_dp_14,signal_dp_15,signal_dp_16,signal_dp_17,signal_dp_18,signal_dp_19,signal_dp_20,signal_dp_21,signal_dp_22,signal_dp_23,signal_dp_24]
images_list = [img_dp_1,img_dp_2,img_dp_3,img_dp_4,img_dp_5,img_dp_6,img_dp_7,img_dp_8,img_dp_9,img_dp_10,img_dp_11,img_dp_12,img_dp_13,img_dp_14,img_dp_15,img_dp_16,img_dp_17,img_dp_18,img_dp_19,img_dp_20,img_dp_21,img_dp_22,img_dp_23,img_dp_24]
freqs_list =[freqs_dp_1,freqs_dp_2,freqs_dp_3,freqs_dp_4,freqs_dp_5,freqs_dp_6,freqs_dp_7,freqs_dp_8,freqs_dp_9,freqs_dp_10,freqs_dp_11,freqs_dp_12,freqs_dp_13,freqs_dp_14,freqs_dp_15,freqs_dp_16,freqs_dp_17,freqs_dp_18,freqs_dp_19,freqs_dp_20,freqs_dp_21,freqs_dp_22,freqs_dp_23,freqs_dp_24] 

#signal_list = [signal_dp_15]
#images_list = [img_dp_15]
#freqs_list =[freqs_dp_15] 

for l in range(0,len(signal_list)):
    signal_path = signal_list[l]
    print("---------------------",signal_path)  
    signal_trackdata = io.loadmat(signal_path)
    signal_data = signal_trackdata['saved_data'];
    
    images_path = images_list[l]
    freqs_path = freqs_list[l]
  
    if (signal_path==signal_o_7 or signal_path==signal_c_2): # These 2 groups belonge to the same date need spesial treatment have different sonication frequency 
        step = 26666
        peaks_frames_locations = [i for i in np.arange(6800,len(signal_data[0]),step)]
        peaks_frames_locations.pop()
        peaks_frames_locations.pop()
        peaks_frames_locations.pop()
        peaks_frames_locations.pop()  
    elif (signal_path==signal_dp_4):
        step = 80000
        peaks_frames_locations = [i for i in np.arange(6800,len(signal_data[0]),step)]
        for s in range(0,26):
            peaks_frames_locations.pop()
    elif (signal_path==signal_dp_5 or signal_path== signal_dp_6 or signal_path== signal_dp_7):
        step = 80000
        peaks_frames_locations = [i for i in np.arange(6800,len(signal_data[0]),step)]
        for s in range(0,6):
            peaks_frames_locations.pop()
    elif (signal_path==signal_dp_8 or signal_path==signal_dp_9 or signal_path==signal_dp_10 or signal_path==signal_dp_11 or signal_path==signal_dp_18 or signal_path==signal_dp_21):
        step = 80000
        peaks_frames_locations = [i for i in np.arange(6800,len(signal_data[0]),step)]
        peaks_frames_locations.pop()        
        peaks_frames_locations.pop()    
    else:
        step = 80000
        peaks_frames_locations = [i for i in np.arange(6800,len(signal_data[0]),step)]
        peaks_frames_locations.pop()
        
    peaks_frames_locations = np.array(peaks_frames_locations) 
    peaks_frames_amplitude = np.ones((len(peaks_frames_locations)))
    
    if(signal_path==signal_o_8 or signal_path==signal_c_5 or signal_path==signal_c_3 or signal_path==signal_c_4):
        widths = np.arange(5, 205) 
        bandwidth = 0.8
        center_frequency = 1.5
        frame_range = 4000
        sub_frame_start = 400
        sub_frame_end = 1900
    else:
        widths = np.arange(5, 205) 
        bandwidth = 1.5
        center_frequency = 1.5 #3.0
        frame_range = 4000
        sub_frame_start = 500 #500
        sub_frame_end = 2500 #1500

    wavelet_frames = []
    for burst_pivot in tqdm(range (0,len(signal_data))):
        burst_signal = signal_data[burst_pivot]
        #Plot_Time_Signal_with_Spikes(burst_signal,peaks_frames_locations,peaks_frames_amplitude) # <---------------------------------------- PLOT 
        #------------------------------------------- Create list of frames Time & Wavelet Domain
        burst_wavelet_frames  = []
        for i in (range(0,len(peaks_frames_locations))): 
            left_frame_side = (peaks_frames_locations[i])
            right_frame_side = (peaks_frames_locations[i]+frame_range)
            if(left_frame_side<0): #this is because some time the first spike is could be between 0 and 1000 not 1000+
                frame_data = burst_signal[0:right_frame_side] 
            elif(right_frame_side>len(burst_signal)): #this is because some time the last spike which have distance less than 1000 from the end.
                frame_data = burst_signal[left_frame_side:]
            else:
                frame_data = burst_signal[left_frame_side:right_frame_side]
            #--------------------------------------------------- CWT calculations 
            sample_Rate = 100000000  # in Hertz <-- (100MHz)
            period = 1/sample_Rate # in seconds
            period = period * 1000000 # in microseconds (μs)
            sample_Rate = sample_Rate / 1000000 # in MegaHz
            cwtmatr, freqs = pywt.cwt(data=frame_data, scales=widths, wavelet='cmor'+str(bandwidth)+'-'+str(center_frequency), sampling_period=1/sample_Rate, method='fft')
            cwt_values = numpy.abs(cwtmatr)
            #Plot_CWT_Small(cwt_values)
            sub_wavelet_frame = cwt_values[:,sub_frame_start:sub_frame_end]  #--------------------------------------------------- Selects a smaller frame // o2=200:1400
            burst_wavelet_frames.append(sub_wavelet_frame)
        burst_wavelet_frames = np.array(burst_wavelet_frames)
        
        resized_burst_wavelet_frames = Resize_list_of_Imges(burst_wavelet_frames,burst_wavelet_frames.shape[1],burst_wavelet_frames.shape[2]/4) #--------------------------------------------------- Compress to smaller frame
        cuted_resized_burst_wavelet_frames = Cut_Values_Outside_Bandwidth_3D(resized_burst_wavelet_frames) #--------------------------------------------------- Cat values outside [0,1]
        wavelet_frames.append(cuted_resized_burst_wavelet_frames)
    wavelet_frames = np.array(wavelet_frames)
    print("Wavelet Frames shape",wavelet_frames.shape)

    #--------------------------------------------------- Save CWT Data
    io.savemat(images_path, mdict={'saved_data': wavelet_frames})
    print("SAVED Wavelets----------------------------------------")
    io.savemat(freqs_path, mdict={'saved_data': freqs})
    print("SAVED Frequences----------------------------------------")
```
