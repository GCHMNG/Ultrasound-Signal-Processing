```python
def Plot_Classes_Properties(classes_labels,classes_center_distances):
    plt.figure(1, figsize=(12,10))
    plt.clf()
    ax = plt.subplot(221) #---------------------------------- Plot Classes over time 
    plt.plot(classes_labels,'o')
    plt.title("Distribution of classified Wavelets during the experiment")
    plt.ylabel("Classes")
    plt.xlabel("Wavelets/Spikes over time")
    plt.show
    ax = plt.subplot(222) #---------------------------------- Plot Adjacency Matrix of Centers
    plt.imshow(classes_center_distances, cmap='seismic', aspect='auto')
    cbar = plt.colorbar()
    plt.title("Distances Between the Centers of each Class")
    plt.ylabel("Classes Centers")
    plt.xlabel("Classes Centers")
    cbar.set_label('Distance of Classes Centers (1=furthest, 0=same spot)',size=10)
    plt.show()
    
#---------------------------- These functions we plot 1 CWT wavelet AND it is called only from the Plot_CTW_Rows_From_Saved_Wavelets. 
def Sub_Plot_Wavelet (ax,i,wavelet,freqs):
    plt.title("Class "+np.str(i))
    sample_Rate = 100000000  # in Hertz <-- (100MHz)
    period = 1/sample_Rate # in seconds
    period = period * 1000000 # in microseconds (μs)
    sample_Rate = sample_Rate / 1000000 # in MegaHz
    #ax = plt.gca()
    im = ax.imshow(wavelet, cmap='seismic', aspect='auto',vmin=0, vmax=1)  # seismic gray
    plt.colorbar(im)
    #---------------------------- Axis
    x_values = np.arange(0,200,35)
    y_values = np.arange(0,len(freqs),15)
    x_labels = []
    for i in x_values:
        x_labels.append(np.str("%.2f" % np.round_(i*period, 4)))
    #y_labels = []
    #for i in y_values:
    #    y_labels.append(np.str("%.2f" % np.round_(freqs[i], 4)))
    ax.set_xticks(x_values)
    ax.set_xticklabels(x_labels)
    ax.set_yticks(y_values)
    #ax.set_yticklabels( y_labels )
    plt.ylabel("Frequency - MegaHertz (MHz)")
    plt.xlabel("Time - microseconds (μs)")

#---------------------------- This function we plot rows of saved wavelets 
def Plot_Rows_of_Wavelets(wavelet_data,freqs,rows_num):
    print("Wavelets Representatives of each Class:")
    wavelets_num = len(wavelet_data)
    i = 0
    while (i < len(wavelet_data)):
        plt.figure(1, figsize=(40,6))
        plt.clf()
        for j in range(0,rows_num):
            if(i+j<wavelets_num):
                ax = plt.subplot(201+rows_num*10+j) #---------------------------------- PLOT
                Sub_Plot_Wavelet (ax,i+j,wavelet_data[i+j],freqs)
                plt.show
            else:
                break
        plt.show()
        i +=rows_num
```
